```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

df = pd.read_csv('f-01.csv',index_col=0)
df.drop(columns=['volume','money','median','hash_close','action'],inplace=True)
df = df.dropna()

# adding features
# df['future_c1'] = round((df['close'].shift(periods=-1) - df['close']) / df['close'],3)
# df['future_c2'] = round((df['close'].shift(periods=-2) - df['close']) / df['close'],3)
# df['future_c3'] = round((df['close'].shift(periods=-3) - df['close']) / df['close'],3)
# df['future_c4'] = round((df['close'].shift(periods=-3) - df['close']) / df['close'],3)


# 策略组1 - 5日线上买绿柱 胜率 70% 持仓3天

df = df[df.today_c>-0.08]
# df = df[df.prev_c1>0.00]
# df = df[df.prev_c2>0.00]
df = df[df.p100_pos>0.4] #越大越好
# df = df[df.mode_pos>0.50]
df = df[df.mode_pos==1]
df = df[df.ma60_pos<0.3]

# 策略组2 - 5日线下买三连绿 持仓1天
# df = df[df.ma5_pos<0.00]
# df = df[df.today_c<0]
# df = df[df.prev_c1<0]
# df = df[df.prev_c2<0]


# print(df.columns)

basefund = 1
error = 0
measure = 'future_c3'
for i in range(len(df)-1):
    change = df[measure].iloc[i]
    if change<-0.01: error+=1
    basefund *= (1+change)

print("-"*100)
print('basefund:',basefund)
print('total:',len(df))
print('errors:',error)
print('errors rate:', round(error/len(df)*100,2))
print("-"*100)

res = df[['security',measure,'ma5_pos','prev_c2','prev_c1','today_c','ma5_ma10','ma60_pos','p100_pos','mode_pos', 'ma10_ang', 'ma60_ang']]
res = res.sort_values(by=[measure],ascending=True)
res[:50]
```
